```python
import math
import numpy as np
data = []
def pi_approx(N):
  t = 0
  f = 0
  for i in range(0,N):
    c1 = np.random.uniform()
    c2 = np.random.uniform()
    if math.sqrt(c1**2 + c2**2) < 1:
      t += 1
      data.append(0)
    else:
      f += 1
      data.append(1)
  return 4*t/(t+f)

pi_approx(20000)
```

```python
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

x = np.linspace(10000,100000,100).astype(int)
y = np.array([pi_approx(i) for i in x])
pi = np.array([math.pi for i in x])

plt.plot(x,y, color = 'r')
plt.plot(x,pi, color = 'b')
plt.show()

print(mean_absolute_error(y,pi))
```
